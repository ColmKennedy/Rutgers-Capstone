## Plots of Airbnb Data

This notebook contains some of the most meaningful plots I found in the Airbnb listings provided to me for this project.

```python
bnb = pd.read_csv('data/airbnb_eda.csv')
# importing data
```

### Interesting Plots

#### Map of New York City's Five Boroughs Using Latitude and Longitude

```python
sns.scatterplot(x = 'longitude',
                y ='latitude',
                hue='neighbourhood_group',
                data=bnb)
plt.show()
```

By plotting the latitude and longitude, we get a relatively accurate map of New York City's five boroughs. This plot does not provide much information about the target variable, price, but verifies the values of latitude, longitude, and neighbourhood_group are as expected.

#### Price by Longitude

```python
sns.scatterplot(x = 'longitude',
                y ='price',
                hue='neighbourhood_group',
                data=bnb)
plt.show()
```

In this set of listings, it appears the most expensive apartments lie in the central region of the city, between about 74 and 73.9 degrees of longitude. Brooklyn and Manhattan cover most of this region, and appear to have a number of more expensive rooms than the rest of the boroughs.

#### Price by Borough and Room Type

```python
sns.barplot(x = 'neighbourhood_group',
            y = 'price',
            hue='room_type',
            data=bnb)
plt.show()
```

Similar to the observations in the previous Manhattan and Brooklyn have more expensive apartments. Although not as obvious in the previous plot, Staten Island also has relatively expenisve apartments. This plot also demonstrates the extreme difference in price based on room type. 

#### Price Against Noise Level of Room

```python
sns.scatterplot(x = 'noise(dB)',
                y ='price',
                hue='neighbourhood_group',
                data=bnb)
plt.show()
```

There appears to be a very linear relationship between noise and price. This relationship appears to have different directions and strengths depending on the Borough it is listed in. 

#### Price Against Floor Number

```python
sns.scatterplot(x = 'floor',
                y ='price',
                hue='neighbourhood_group',
                data=bnb)
plt.show()
```

Similar to the previous plot, there appears to be distinct linear relationships by Borough. In other words, depending on the Borough, the floor number has more or less power on the price.

#### Floor vs. Noise Level

```python
sns.scatterplot(x = 'noise(dB)',
                y ='floor',
                hue='neighbourhood_group',
                data=bnb)
plt.show()
```

When examining floor and noise, we can see a correlation between floor number and noise level. This is concerning for modeling as it could be an indicator of collinearity, therefore creating potential for the model to overfit.
